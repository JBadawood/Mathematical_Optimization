#Part (a)
**Question:** What is the probability that less than *215 customers* will enter the shop between 2:30 p.m. and 8:15 p.m.?

```python
lambda_345=345/1.7
```

```python
poisson.cdf(214, lambda_345)
```

#Part (c)
**Question:** What is the probability that at least *850 meals* will be ordered during a day?

```python
from scipy.stats import norm
```

```python
lambda_day=630/1.7
```

```python
lambda_day
```

```python
expected_meals_per_customer = 1*0.48 +2*0.27 + 4*0.09 + 5*0.13 + 7*0.03
```

```python
expected_meals_per_customer
```

```python
expected_meals_per_day = lambda_day * expected_meals_per_customer
```

```python
expected_meals_per_day
```

```python
# convert poisson to normal approximation
mu=expected_meals_per_day
sigma = (expected_meals_per_day)**0.5
```

```python
norm.sf(850,mu,sigma)
```
