#### Import NumPy pakage

```python
# أول شي نحتاج نثبت مكتبة NumPy
import numpy as np
```

```python
# بهذه الطريقة أقدر أنشئ array بالقيم إلى أبغاها
data = np.array([1,2,3])
```

```python
np.array([1,2,3])
```

```python
# عشان أحصل على أعلى قيمة في الـ array السابق
data.max()
```

```python
# لو نبغى كل الثلاث قيم بواحد
ones = np.ones(3)
ones
```

```python
# لو نبغاها بصفر راح نكتب التالي
np.zeros(3)
```

```python
# أما إذا حابين نخليها 3 قيم عشوائية
np.random.random(3)
```

```python
# كذلك نقدر ننشئ وحدة بدون ما نحدد ايش الأرقام مبدئيًا
np.empty(3)
```

```python
modified_data = data + ones
modified_data
```

```python
# عشان أظهر وحدة من عناصر الـ array
modified_data[0]
```

```python
# عشان أظهر جزء من البيانات وأعمل slice
modified_data[0:2]
```

#### 1-D Array

```python
np.array([0,1,2])
```

 نقدر نعمل الـ array بطريقة ثانية

```python
np.arange(3)
```

```python
# نقدر نحدد البداية والنهاية وكم يقفز العد
np.arange(1,10,2)
```

طريقة ثالثة بتحديد عدد العناصر في الـ array

```python
# نحدد البداية والنهاية وكم عنصر
np.linspace(0,10,5)
```

```python
# نقدر نشيك على نوع البيانات
(np.linspace(0,10,5)).dtype
```

> نلاحظ أن نوع البيانات في الـ array السابق هو float  
نقدر نغيره لأعداد صحيحة

#### 2-D & Multidimensional Arrays (ndarray)

```python
# 2-D array
np.array([[1,2,3],[4,5,6]])
```

```python
# نقد نعيد تشكيل الـ one-dimensional array
one_D = np.array([1,2,3,4,5,6])

two_D = np.reshape(one_D, (2,3))
```

```python
two_D
```

```python
# عشان نشوف الـ array هو كم بعد
two_D.ndim
```

```python
# ولمعرفة عدد صفوف المصفوفة وعدد أعمدتها
two_D.shape
```

```python
# أما إذا نبغى نعرف عدد عناصر المصفوفة
two_D.size
```
